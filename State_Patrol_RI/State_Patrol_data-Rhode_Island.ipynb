# Feature Engineering

## Checking the categorical features

```python
continous=[]
categorical=[]
def convert_type(a):
    n=df[a].nunique()
    if n > 20:
        continous.append(a)
    else:
        categorical.append(a)
```

```python
for index,i in enumerate(df.columns):
    convert_type(i)
```

```python
categorical
```

```python
df[categorical].iloc[-20:]
```

```python
def print_unique(a):
    for i in a:
        print(df[i].unique())
```

```python
print_unique(df[categorical])
```

```python
# I will replace the NaNs with 0 and make a new category out of it or using my domain knowedge i will replace some with mode for that particular features
```
